```python
# load the three sector_data csv files into pandas dataframes
it_sector_data = pd.read_csv('information_technology_sector_data.csv', sep=',', header = 0)
hc_sector_data = pd.read_csv('health_care_sector_data.csv', sep=',', header = 0)
cd_sector_data = pd.read_csv('consumer_discretionary_sector_data.csv', sep=',', header = 0)
```

```python
# for each of the three rename Open, High, Low, Close, Adj Close, and Volume columns to include the sector name
it_sector_data.rename(columns={'Open': 'IT Open', 'High': 'IT High', 'Low': 'IT Low', 'Close': 'IT Close', 'Adj Close': 'IT Adj Close', 'Volume': 'IT Volume'}, inplace=True)
hc_sector_data.rename(columns={'Open': 'HC Open', 'High': 'HC High', 'Low': 'HC Low', 'Close': 'HC Close', 'Adj Close': 'HC Adj Close', 'Volume': 'HC Volume'}, inplace=True)
cd_sector_data.rename(columns={'Open': 'CD Open', 'High': 'CD High', 'Low': 'CD Low', 'Close': 'CD Close', 'Adj Close': 'CD Adj Close', 'Volume': 'CD Volume'}, inplace=True)
```

```python
# merge the three sector dfs into one dataframe
merged_sector_data = pd.merge(it_sector_data, hc_sector_data, on='Date')
merged_sector_data = pd.merge(merged_sector_data, cd_sector_data, on='Date') 
```

Loading daily WTI prices:

The csv file has been downloaded from OK Cushing WTI Spot Prices from the U.S. Energy Information Administration

```python
# loading daily oil prices
oil_prices = pd.read_csv('WTI_oil_daily.csv', sep=';', decimal = ',', header = 0) 
```

```python
# Change Date column to datetime
oil_prices['Date'] = pd.to_datetime(oil_prices['Date'])
```

```python
oil_prices
```

```python
# change Date column for oil_prices to datetime
oil_prices['Date'] = pd.to_datetime(oil_prices['Date'])
# change Date column for merged_sector_data to datetime
merged_sector_data['Date'] = pd.to_datetime(merged_sector_data['Date'])
```

Merging the WTI and Sector dataframes

```python
# inner merge on oil_prices and  from the 3 sectors
merged_data = pd.merge(oil_prices, merged_sector_data, on='Date', how='inner')
```

```python
merged_data.head()
```

Importing daily VIX data from the csv, which was downloaded directly from CBOE

```python
# import VIX_History.csv
vix_data = pd.read_csv('VIX_History.csv', sep=',', header = 0)
```

```python
# Rename DATE to Date, and OPEN to VIX Open, HIGH to VIX High, LOW to VIX Low, CLOSE to VIX Close
vix_data.rename(columns={'DATE': 'Date', 'OPEN': 'VIX Open', 'HIGH': 'VIX High', 'LOW': 'VIX Low', 'CLOSE': 'VIX Close'}, inplace=True)

# change Date column for vix_data to datetime
vix_data['Date'] = pd.to_datetime(vix_data['Date'])
```

```python
vix_data
```

```python
# merge vix_data with merged_data
final_data = pd.merge(merged_data, vix_data, on='Date', how='inner')
```

```python
# save as CompleteData.csv
final_data.to_csv('CompleteData.csv', index=False)
```

### Some Feature Engineering

```python
import pandas as pd
import numpy as np

# Load dataset
file_path = "CompleteData.csv"  # Adjust the path if needed
df = pd.read_csv(file_path, parse_dates=['Date'])

# Sort by date just in case
df = df.sort_values(by="Date")

# Selecting relevant columns
selected_columns = [
    "Date", "WTI Spot Price", "VIX Close", 
    "IT Adj Close", "HC Adj Close", "CD Adj Close",
    "IT Volume", "HC Volume", "CD Volume"
]

df = df[selected_columns]

# Calculate Log Returns for each sector
for col in ["IT Adj Close", "HC Adj Close", "CD Adj Close"]:
    df[f"{col[:2]}_Log_Return"] = np.log(df[col] / df[col].shift(1))

# Calculate Log Percentage Change for WTI Spot Price
df["WTI Log Daily Change"] = np.log(df["WTI Spot Price"] / df["WTI Spot Price"].shift(1))

# Drop rows with NaN values due to shifting operations
df.dropna(inplace=True)
```

```python
import pandas as pd
import numpy as np

# Load dataset
file_path = "CompleteData.csv"  # Adjust the path if needed
df = pd.read_csv(file_path, parse_dates=['Date'])

# Sort by date just in case
df = df.sort_values(by="Date")

# Selecting relevant columns
selected_columns = [
    "Date", "WTI Spot Price", "VIX Close", 
    "IT Adj Close", "HC Adj Close", "CD Adj Close",
    "IT Volume", "HC Volume", "CD Volume"
]

df = df[selected_columns]

# Calculate the returns for each sector
for col in ["IT Adj Close", "HC Adj Close", "CD Adj Close"]:
    df[f"{col[:2]}_Return"] = df[col].pct_change()

# Calculate Percentage Change for WTI Spot Price
df["WTI Daily Change"] = df["WTI Spot Price"].pct_change()

# Drop rows with NaN values due to shifting operations
df.dropna(inplace=True)
```

```python
# save as ForecastData.csv
df.to_csv('ForecastData.csv', index=False)
```
