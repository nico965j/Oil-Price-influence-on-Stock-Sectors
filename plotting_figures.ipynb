# Plotting figures from literary review myself

```python
# loading libraries
#import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## AI and ML in finance literature growth

```python
# y-axis is Number of documents, x-axis is years starting from 2011 to 2021. The numbers in question are 6,2,8,9,13,16,16,29,52,84,112
# creating a list of years
years = list(range(2011,2022))
# creating a list of numbers
numbers = [6,2,8,9,13,16,16,29,52,84,112]


# Creating a professional-looking plot
plt.figure(figsize=(12, 6))

# Smoother line with markers
plt.plot(years, numbers, marker='o', linestyle='-', color='navy', linewidth=2, 
         markersize=6, markerfacecolor='orange', markeredgewidth=1, markeredgecolor='black')

# Labels and title with improved font styles
plt.xlabel('Years', fontsize=14,  labelpad=10)
plt.ylabel('Number of Documents', fontsize=14, labelpad=10)
plt.title('Literature Growth: AI & ML in Finance', fontsize=16, pad=20)

# Grid styling for readability
plt.grid(True, linestyle='--', linewidth=0.6, alpha=0.7)

# Formatting x and y ticks
plt.xticks(years, rotation=45, fontsize=12)
plt.yticks(np.arange(0, max(numbers) + 10, 10), fontsize=12)

# Adjusting layout for better spacing
plt.tight_layout()

# Display the plot
plt.show()

```

## Annual publications trend

```python
import matplotlib.pyplot as plt
import numpy as np

# Example Data (Replace with actual data)
years = np.arange(2008, 2022)  
publications = [5, 5, 10, 11, 11, 25, 26, 33, 51, 99, 101, 125, 135, 43]  # Example values

# Creating a professional bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(years, publications, color='royalblue', edgecolor='black', alpha=0.8)

# Annotate values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 3, f'{int(yval)}', ha='center', fontsize=12)

# Labels and title
plt.xlabel('Year', fontsize=14, labelpad=10)
plt.ylabel('Number of Publications', fontsize=14, labelpad=10)
plt.title('Annual Publications Trend', fontsize=16, pad=20)

# Grid styling
#plt.grid(axis='y', linestyle='--', linewidth=0.6, alpha=0.7)

# Formatting x and y ticks
plt.xticks(years, rotation=45, fontsize=12)
plt.yticks(np.arange(0, max(publications) + 30, 20), fontsize=12)

# Adjust layout for spacing
plt.tight_layout()

# Display the plot
plt.show()
```
